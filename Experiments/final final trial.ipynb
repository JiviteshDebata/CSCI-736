```python
import pandas as pd
import numpy as np
import sys
import schedule
import time
```

```python
train_data = pd.read_csv(r"C:\Users\johns\Documents\NNandML\OpenSet\Project\CSCI-736\Experiments\mnist_train.csv")
test_data = pd.read_csv(r"C:\Users\johns\Documents\NNandML\OpenSet\Project\CSCI-736\Experiments\mnist_test.csv")

# train_data.head()
test_data.head()
```

```python
training_features=train_data.drop('label',axis=1)

training_features.head()
```

```python
training_labels=train_data['label']

training_labels.head()
```

```python
testing_features=test_data.drop('label',axis=1)

testing_features.head()
```

```python
testing_labels=test_data['label']

testing_labels.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_training_features = scaler.fit_transform(training_features)
scaled_testing_features = scaler.transform(testing_features)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(scaled_training_features, training_labels)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
original_predictions = model.predict(scaled_testing_features)
print(confusion_matrix(testing_labels, original_predictions))
print(classification_report(testing_labels, original_predictions))
```

```python


original_predictions = model.predict(scaler.transform(testing_features))

modified_testing_features = testing_features.copy()
# start_time = datetime.datetime.now().time()
# Iterate over each image index in the original_predictions array
for image_index, original_prediction in np.ndenumerate(original_predictions):
    # Modify each pixel of the image one by one and test it
    for row in range(28):
        for col in range(28):
            pixel_index = row * 28 + col

            # Create a temporary copy of the image with the modified pixel value
            temp_modified_image = modified_testing_features.iloc[image_index[0]]
            temp_modified_image[pixel_index] = temp_modified_image[pixel_index] + 1
            

            # Scale the modified image
            scaled_temp_modified_image = scaler.transform([temp_modified_image])

            # Test the modified image
            modified_prediction = model.predict(scaled_temp_modified_image)
            print(original_prediction, modified_prediction)
            if original_prediction != modified_prediction:
                # save original pic
                # save modified pic
                temp_modified_image.to_csv("Modified_image.csv")
                testing_features.iloc[image_index[0]].to_csv("Original_image.csv")
                print( "Modifying pixel: Row: ", row, ", Column:", col, "in image: ", image_index, "to pixel_val+1 changes the prediction from: ", original_prediction, "to: ", modified_prediction )
                sys.exit(0)

            temp_modified_image = modified_testing_features.iloc[image_index[0]]
            temp_modified_image[pixel_index] = temp_modified_image[pixel_index] - 1
            

            # Scale the modified image
            scaled_temp_modified_image = scaler.transform([temp_modified_image])

            # Test the modified image
            modified_prediction = model.predict(scaled_temp_modified_image)
            print(original_prediction, modified_prediction)
            if original_prediction != modified_prediction:
                # save original pic
                # save modified pic
                temp_modified_image.to_csv("Modified_image.csv")
                testing_features.iloc[image_index[0]].to_csv("Original_image.csv")
                print( "Modifying pixel: Row: ", row, ", Column:", col, "in image: ", image_index, "to pixel_val-1 changes the prediction from: ", original_prediction, "to: ", modified_prediction )
                sys.exit(0)
            schedule.run_pending()
            

```
